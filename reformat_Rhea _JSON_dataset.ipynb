```python
import pandas as pd
from google.colab import drive
import os
import csv
import json
drive.mount('/content/drive')
```

```python
with open('/content/drive/MyDrive/datasets/positive.json', 'r') as f:
    data = json.load(f)
```

```python
for i in range(len(data)):
  if "," in data[i][0]:
    data[i] = data[i][0].split(",")
```

```python
pos = pd.DataFrame(data,columns= ['Protein_ID', 'Reaction_ID'])
pos.head()
```

```python
with open('/content/drive/MyDrive/datasets/negative.json', 'r') as f:
    data = json.load(f)
```

```python
for i in range(len(data)):
  if "," in data[i][0]:
    data[i] = data[i][0].split(",")
```

```python
neg = pd.DataFrame(data,columns= ['Protein_ID', 'Reaction_ID'])
neg.head()
```

```python
with open('/content/drive/MyDrive/datasets/uniprot_dict.json', 'r') as f:
    data = json.load(f)
```

```python
uniprot_dict = pd.DataFrame(list(data.items()), columns=['Protein_ID', 'Sequence'])
uniprot_dict.head()
```

```python
with open('/content/drive/MyDrive/datasets/rhea_dict.json', 'r') as f:
    data = json.load(f)
```

```python
rhea_dict = pd.DataFrame(list(data.items()), columns=['Reaction_ID', 'SMILES'])

rhea_dict.head()
```

```python
merged_df = pd.merge(pos , uniprot_dict, on='Protein_ID')
merged_df = pd.merge(merged_df, rhea_dict, on='Reaction_ID')

Positives = merged_df[['SMILES', 'Sequence']]
Positives.head()
```

```python
Positives.to_json('/content/drive/MyDrive/datasets/Pos_rxn_seq.json', orient='records', lines=True)
```

```python
merged_df = pd.merge(neg , uniprot_dict, on='Protein_ID')
merged_df = pd.merge(merged_df, rhea_dict, on='Reaction_ID')

Negatives = merged_df[['SMILES', 'Sequence']]
Negatives.head()
```

```python
Negatives.to_json('/content/drive/MyDrive/datasets/Neg_rxn_seq.json', orient='records', lines=True)
```
